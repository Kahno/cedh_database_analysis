```python
import pandas as pd
import numpy as np
import json

from sklearn.cluster import KMeans
from analyze import *
```

```python
dataset = load_database()
all_cards, num_decks = summary(dataset)
all_cards_x = list(filter(not_basic_land, all_cards.keys()))
```

```python
len(all_cards_x)
```

```python
dataset_flat = {}

for color in dataset:
    for deck_type in dataset[color]:
        for deck in dataset[color][deck_type]:
            dataset_flat[deck] = dataset[color][deck_type][deck]
            
deck_names = list(sorted(dataset_flat.keys()))
all_cards_dict = {card: i for i, card in enumerate(all_cards_x)}
mat = []

for i, deck in enumerate(deck_names):
    cur_line = [0] * len(all_cards_x)
    for card in dataset_flat[deck]:
        if not_basic_land(card):
            cur_line[all_cards_dict[card]] = 1
    mat.append(cur_line)

df = pd.DataFrame(
    data=np.array(mat), 
    index=deck_names, 
    columns=all_cards_x
)
```

```python
df
```

```python
def similarity(a, b, output=False):
    v = df.loc[a].values
    w = df.loc[b].values
    
    result = sum(v & w) / sum(v | w)
    
    if output:
        print(f"Decks:\n\t- {a}\n\t- {b}\n")
        print(f"{sum(v & w)} card(s) in common")
        print(f"Jaccard similarity: {result}")
    
    return result
```

```python
similarity("Zur Doomsday", "Bluepod")
```

```python
similarity("Zur Doomsday", "Zur Doomsday")
```

```python
top_sim = 0
bot_sim = 1

most_sim = []
least_sim = []

for deck1 in deck_names:
    for deck2 in deck_names:
        if deck1 == deck2:
            continue
            
        cur_sim = similarity(deck1, deck2)
        if cur_sim > top_sim: 
            top_sim = cur_sim
            most_sim = [deck1, deck2]
            
        if cur_sim < bot_sim:
            bot_sim = cur_sim
            least_sim = [deck1, deck2]
            
            
print(f"Most similar decks are {most_sim[0]} and {most_sim[1]}: {similarity(*most_sim)}")
print(f"Least similar decks are {least_sim[0]} and {least_sim[1]}: {similarity(*least_sim)}")
```
